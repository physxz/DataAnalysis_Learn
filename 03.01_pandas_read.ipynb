```python
temp_actors_list = df['Actors'].str.split(', ').to_list()
temp_actors_list
```

```python
actors_list = [i for j in temp_actors_list for i in j]
actors_list
```

```python
actors_num = len(set(actors_list)) # 去重演员个数
actors_num
```

```python
runtime_data = df['Runtime (Minutes)'].values
runtime_max = runtime_data.max()
runtime_min = runtime_data.min()
num_bins = ( runtime_max - runtime_min ) // 10

fig, ax = plt.subplots(figsize=(10,3), dpi=100)
ax.hist(runtime_data, num_bins)

ax.set_xticks(np.arange(runtime_min, runtime_max+10, 10))
```

```python
rating_data = df['Rating'].values
rating_max = rating_data.max()
rating_min = rating_data.min()
num_bins = int( ( rating_max - rating_min ) // 0.5 )

fig, ax = plt.subplots(figsize=(10,3), dpi=100)
ax.hist(rating_data, num_bins)

ax.set_xticks(np.arange(rating_min, rating_max+0.5, 0.5))
```
